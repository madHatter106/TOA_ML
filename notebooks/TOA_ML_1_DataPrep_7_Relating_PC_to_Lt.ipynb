```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as pl
from seaborn import heatmap
from cmocean.cm import balance_r

from latexify import latexify
```

```python
latexify()
```

```python
ds = pd.read_pickle('./pickleJar/OperationalDataSets/df_6_APHY_Standardized_SWF.pkl')
```

```python
dsat = pd.read_pickle('./pickleJar/OperationalDataSets/df_6_APHY_Standardized_SWF_SAT.pkl')
```

```python
dsat.head().T
```

```python
dsat.reset_index(inplace=True)
```

```python
dsat.head().T
```

```python
ds.info()
```

```python
ds.head().T
```

```python
d_sat_pc = pd.concat((#dsat[['log10_etopo2_s', 'oisst_s', 'solz_s']],
                      dsat.filter(regex='rho', axis=1),
                      ds.filter(regex='PC', axis=1)), axis=1)
```

```python
d_sat_pc.head()
```

```python
d_sat_pc_corr = d_sat_pc.corr().iloc[1:, :-1]
```

```python
d_sat_pc_corr
```

```python
ticklabels = [r'$log(depth)$', r'$SST$', r'$Sol. Zen.$'] +\
                [r'$Lt_{rc}$(%d)' % i for i in [411, 443, 489, 510, 555, 670]] +\
                ['PC%d' % i for i in range(1, 7)]
```

```python
ticklabels = [r'$Lt_{rc}$(%d)' % i for i in [411, 443, 489, 510, 555, 670]] +\
                ['PC%d' % i for i in range(1, 7)]
```

```python
latexify(fig_width=6, fig_height=5)
```

```python
f, ax = pl.subplots(figsize=(6, 5))
ax.set_yticklabels(ticklabels[1:])
ax.set_xticklabels(ticklabels[:-1])
heatmap(d_sat_pc_corr, vmin=-1, vmax=1, cmap=balance_r, annot=True,
       mask = np.triu(np.ones(d_sat_pc_corr.shape), k=1), ax=ax,
        xticklabels=ticklabels[:-1], yticklabels=ticklabels[1:]
       #annot_kws={'fontsize':14}
       )
ax.set_facecolor('k')
#ax.axvline(, color='k', linewidth=2)
ax.axvline(6, color='k', linewidth=2)
ax.axhline(5, color='k', linewidth=2)
f.savefig('./figJar/latexified_input_heatmap.pdf', dpi=300, format='pdf')
```
